### Import Packages

```python
import pandas as pd
import numpy as np
import dask.dataframe as dd
import dask.array as da
import dask.bag as dbs
```

### Import Data

```python
#creates dask dataframe
ddf = dd.read_csv('./data/Iowa_Liquor_Sales.csv', dtype = {'Category': 'float64',
                                                    'County Number': 'float64',
                                                    'Zip Code': 'object',
                                                    'Vendor Number': 'float64',
                                                    'Item Number': 'object'},
                                            blocksize = '10MB').reset_index()
```

```python
#check size of file for correct number of rows
ddf.shape[0].compute()
```

```python
#convert to parquet file
ddf.to_parquet('./data/parquet/', name_function = lambda x: f"data-{x}.parquet")
```
